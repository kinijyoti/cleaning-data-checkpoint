```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

jobs_df = pd.read_csv("jobs_in_data.csv")
print(jobs_df.info())
```

## Irregular Data

Detect outliers in your dataset and handle them as needed. Use code comments to make notes about your thought process.

```python
# In the EDA process, I explored the Salary in USD column with visualizations and discovered some outliers. Want to use another method to see what those outliers might be. 

print(jobs_df['salary_in_usd'].describe())


# To find inconsistent or unusual categorical (Checking for unique values and their frequencies) 
# values in fields like job_title, experience_level, and employment_type:

job_title_counts = jobs_df['job_title'].value_counts()
print(job_title_counts)
experience_level_counts = jobs_df['experience_level'].value_counts()
print(experience_level_counts)
employment_type_counts = jobs_df['employment_type'].value_counts()
print(employment_type_counts)

# # Some of the Salaries for the "Data analyst" jobs are high paid salaries for the same expereince level are the outliers filtered the salary by Employee residence, Job title and experience level

outliers_records_in_Data_Analysis_job= jobs_df[(jobs_df['employee_residence'] == 'United States') & (jobs_df['job_title'] == 'Data Analyst') & (jobs_df['experience_level'] == 'Mid-level')& (jobs_df['work_year'] == 2023) ].sort_values(by='salary_in_usd', ascending=False)
print(outliers_records_in_Data_Analysis_job)

# # Some of the Salaries for the "Data Engineer" jobs are high paid salaries for the same expereince level are the outliers filtered the salary by Employee residence, Job title and experience level

outliers_records_in_Data_Engineer_job= jobs_df[(jobs_df['employee_residence'] == 'United States') & (jobs_df['job_title'] == 'Data Engineer') & (jobs_df['experience_level'] == 'Mid-level') & (jobs_df['work_year'] == 2023) ].sort_values(by='salary_in_usd', ascending=False)
print(outliers_records_in_Data_Engineer_job)

# Detect outliers using the IQR method for 'Salary'
Q1 = jobs_df['salary_in_usd'].quantile(0.25)
Q3 = jobs_df['salary_in_usd'].quantile(0.75)
IQR = Q3 - Q1
print(Q1)
print(Q3)
print(IQR)

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(lower_bound)
print(upper_bound)

# Identifying outliers

salary_usd_outliers = jobs_df[(jobs_df['salary_in_usd'] < lower_bound) | (jobs_df['salary_in_usd'] > upper_bound)]
print(salary_usd_outliers)

# There are 158 outliers in the "salary_in_usd" column. Many of these outliers are associated with high-paying roles in machine learning and AI, 
# especially at the senior level, which could indicate genuine high salaries rather than data errors. Hereâ€™s a brief summary:
# Common roles among outliers: Machine Learning Engineer, ML Engineer, Research Scientist. 
# Most outliers are senior-level positions. Work settings are prsented for both remote and in-person. 
# Many high salaries are from the United States and Canada.

```

## Inconsistent Data

Check for inconsistent data and address any that arises. As always, use code comments to illustrate your thought process.

```python
# Group by job title and calculate min, max, and range of salaries
salary_ranges = jobs_df_unique.groupby('job_title').agg({
    'salary_in_usd': ['min', 'max']
})
print(salary_ranges)

# Calculation of the salary range and percentage difference
salary_ranges['salary_range'] = salary_ranges['salary_in_usd']['max'] - salary_ranges['salary_in_usd']['min']
salary_ranges['percentage_difference'] = (salary_ranges['salary_range'] / salary_ranges['salary_in_usd']['min']) * 100
print(salary_ranges)

# Sort by percentage difference in descending order
salary_ranges = salary_ranges.sort_values('percentage_difference', ascending=False)
print(salary_ranges)

# job titles with the largest salary ranges
print(salary_ranges.head(10))

# Identify job titles with unusually large salary ranges (e.g., more than 100% difference)
large_range_jobs = salary_ranges[salary_ranges['percentage_difference'] > 100]
print("\nJob titles with salary ranges exceeding 100% difference:")
print(large_range_jobs)

plt.figure(figsize=(12, 6))
sns.boxplot(x='work_setting', y='salary_in_usd', data=jobs_df_unique)

plt.title('Salary Distribution by Work Setting', fontsize=16)
plt.xlabel('Work Setting', fontsize=12)
plt.ylabel('Salary in USD', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: format(int(x), ',')))
# Show the plot
plt.tight_layout()
plt.show()
```
